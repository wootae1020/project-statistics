# 1. 중복 값을 판매수 다 더하는 것으로 처리

```python
result = df_service_origin.groupby(['서비스명', '판매자','서비스번호'], as_index=False).agg({'판매수': 'sum','서비스가격':'first','평점':'first','대분류':'first'})

# 결과 출력
print(result)
result.to_csv('service_sum.csv', encoding='cp949', index=False)
```
