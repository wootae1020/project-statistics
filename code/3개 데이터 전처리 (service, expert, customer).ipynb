```python
# '카테고리 ' 타입 컬럼에 대한 통계 정보 얻기
category_summary = df.describe(include=['category'])

# 결과 출력
print(category_summary)
```

```python
import pandas as pd

# '카테고리' 타입 컬럼만 선택
category_columns = df.select_dtypes(include=['category'])

# 1. 기본적인 기술통계 정보
category_summary = category_columns.describe(include=['category'])

# 2. 각 컬럼별 고유값 개수
unique_counts = category_columns.nunique()

# 3. 각 컬럼별 결측값 개수
missing_counts = category_columns.isnull().sum()

# 4. 각 컬럼별 최빈값과 최빈값의 개수
mode_values = category_columns.mode().iloc[0]  # 최빈값 (여러 개일 경우 첫 번째 값)
mode_counts = category_columns.apply(lambda x: x.value_counts().iloc[0])  # 최빈값의 개수

# 5. 결과를 하나의 데이터프레임으로 정리
category_stats = pd.DataFrame({
    '고유값 개수': unique_counts,
    '결측값 개수': missing_counts,
    '최빈값': mode_values,
    '최빈값 개수': mode_counts
})

# 결과 출력
print(" 기본 기술통계 정보:\n")
display(category_summary)

print("\n 추가 통계 정보:\n")
display(category_stats)
```
